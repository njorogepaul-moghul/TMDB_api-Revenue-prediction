# 2. 🧹 Data cleaning
- In this notebook we are going to perform data cleaning operations to the `merged_df` dataset.

```python
#lets import the libraries
import pandas as pd
```

```python
#lets load the merged dataset
merged_df=pd.read_csv('C:/Users/Win/TMDB project/data/movies_full.csv')
merged_df.head()
```

```python
merged_df.columns
```

```python
#first we  are going to drop all the x-labeled columns since they appeared because both dataframes had overlapping column names
# they basically have the same info
merged_df=merged_df.drop(columns=['popularity_x', 'release_date_x', 'title_x', 'vote_average_x', 'vote_count_x'])
merged_df.columns
```

```python
#lets noew rename the y_labeled columns to cleaner names
merged_df=merged_df.rename(columns={
    'title_y':'title',
    'popularity_y':'popularity',
    'release_date_y':'release_date',
    'vote_average_y':'vote_average',
    'vote_count_y':'vote_count'
    
})
merged_df.columns
```

```python
#lets see the shape of the dataset
merged_df.shape
print("shape of the dataset",merged_df.shape)
```

```python
#lets get some information on the dataset
merged_df.info()
```

```python
#lets get a statistical description of our dataset
merged_df.describe()
```

```python
#lets check for duplicates
merged_df.duplicated().sum()
```

- **we have 326 duplicated rows in our dataset, we opt to drop them**

```python
#lets drop the duplicates
merged_df.drop_duplicates(inplace=True)
```

```python
#lets confirm there are no more duplicates
merged_df.duplicated().sum()
```

```python
#lets check for null values
merged_df.isnull().sum()
```

```python
merged_df.head(2)
```

```python
#lets drop some columns since they are of no significance in our project
merged_df.drop(columns=['backdrop_path','poster_path','adult', 'video', 'status',
                        'original_title', 'id','genre_ids'],inplace=True)

#WHY WE DROPPED:
#adult: almost all are False, no predictive value
#video: always False for most movies
#status: usually “Released” for all
#overview:long text, unless you plan NLP analysis
#original_title: duplicate of title
#id: identifier only, 
#genre_ids: just a list of numeric lists reperesenting the genres in the genre column
```

```python
merged_df.columns
```

```python
#lets drop the five rows with missing release_date
merged_df.dropna(subset=['release_date'], inplace=True)
```

```python
#lets also drop the rows with missing overview columns
merged_df.dropna(subset=['overview'], inplace=True)
```

```python
merged_df.isnull().sum()
```

- **🙌 now we have no missing values**
- **LETS DEFINE THE COLUMNS**

- original_language → language the movie was produced in (e.g., "en", "fr")

- overview → short plot summary or description of the movie

- title → movie’s official title

- budget → production cost (in USD)

- revenue → total earnings (in USD) — your target variable

- runtime → movie duration in minutes

- release_date → when the movie was released

- popularity → TMDB’s popularity metric based on views and ratings

- vote_average → average user rating (0–10)

- vote_count → total number of user ratings

- genres → list of movie genres (e.g., Action, Drama, Comedy)

```python
#since we can observe that, some row have zero entries in the revenue an budget columns , lets see how many
print("zero entries in the revenue column:",(merged_df['revenue'] == 0).sum())
print("zero entries in the budget column:",(merged_df['budget'] == 0).sum())
```

```python
# we opt to drop these rows for practicality
merged_df = merged_df[(merged_df['budget'] != 0) & (merged_df['revenue'] != 0)]
```

```python
print("The shape of our dataset after cleaning",merged_df.shape)
```

```python
merged_df.head()
```

- **Even though the dataset shrinked,We now have a clean dataset.**

```python
#lets now save the cleaned dataset
merged_df.to_csv('C:/Users/Win/TMDB project/data/cleaned_movies.csv', index=False)
```
